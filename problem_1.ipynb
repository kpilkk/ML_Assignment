```python
import pandas as pd
from scipy.stats import multivariate_normal
# for displaying only first 10 rows and first 10 columns
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
```

By default, `pd.read_csv` uses `header=0` (when the names parameter is also not specified) which means the first (i.e. 0th-indexed) line is interpreted as column names.

If our data has no header, then use
`pd.read_csv(..., header=None)`

```python
data_train = pd.read_csv("P1_data/P1_data_train.csv", header=None)
label_train = pd.read_csv("P1_data/P1_labels_train.csv", header=None)
```

Displaying first 5 labels using `Dataframe.head()`

```python
label_train.head()
```

Creating two list to store the index values of 5 and 6
`index_5` & `index_6` are list which contains row no of label 5 & 6 respectively.

```python
index_5 = []
index_6 = []
for i, j in label_train.iterrows():
    if j[0] == 5:
        index_5.append(i)
    else:
        index_6.append(i)
```

No. of elements which are labelled as 5 in training dataset.

```python
len(index_5)
```

### Individual probability of label being 5 or 6
i.e. `prior probabilities` <br> `index_5_data` and `index_6_data` stores data which are labelled as 5 and 6 respectively.

```python
prob_5 = len(index_5)/len(label_train)
prob_6 = 1 - prob_5
index_5_data = data_train.iloc[index_5]
index_6_data = data_train.iloc[index_6]
```

### Mean of label 5

```python
mean_5 = index_5_data.mean()
mean_5
```

### Meanof label 6

```python
mean_6 = index_6_data.mean()
mean_6
```

### Covariance of label 5

```python
covariance_5 = index_5_data.cov()
covariance_5
```

### Covariance of label 6

```python
covariance_6 = index_6_data.cov()
covariance_6
```

Continuing with the test data

```python
data_test = pd.read_csv("P1_data/P1_data_test.csv", header=None)
data_test
```

Making multivariate_normal pdf of test data given `mean` and `variance` calculated from training data. <br> And calculating posterior probabilties by multiplying with corresponding prior probabilities.

```python
y1 = multivariate_normal.pdf(data_test,mean=mean_5,cov=covariance_5)*prob_5
y2 = multivariate_normal.pdf(data_test,mean=mean_6,cov=covariance_6)*prob_6
```

```python
li = []
for i in range(len(y1)):
    if y1[i] > y2[i]:
        li.append(5)
    else:
        li.append(6)
```

```python
true_label = pd.read_csv('P1_data/P1_labels_test.csv', header = None)
```

```python
count = 0
for i in range(len(true_label)):
    if(true_label[0][i] == li[i]):
        count += 1
        
count
```

It gave correct label of 257 data out of 333. <br>
Now, finding the confusion matrix as we know the correct label and the predicted label.

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(true_label,li)
print(conf_mat)
```

As the misclassification error is found to be 22.8%. Accuracy is 77.2%.

```python
misclassification_error = (conf_mat[0][1]+conf_mat[1][0])/(conf_mat[0][0]+conf_mat[0][1]+conf_mat[1][0]+conf_mat[1][1])
misclassification_error
```
